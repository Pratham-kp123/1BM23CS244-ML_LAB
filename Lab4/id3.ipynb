```python

# DECISION TREE CLASSIFICATION
# Datasets: Iris & Wine


import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay


# IRIS DATASET

iris = load_iris()
X1, y1 = iris.data, iris.target

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.3, random_state=42
)

clf1 = DecisionTreeClassifier(max_depth=3, random_state=42)
clf1.fit(X1_train, y1_train)

y1_pred = clf1.predict(X1_test)

print("IRIS DATASET")
print("Accuracy:", accuracy_score(y1_test, y1_pred))
print("F1 Score (macro):", f1_score(y1_test, y1_pred, average='macro'))

# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y1_test, y1_pred)
plt.title("Iris Confusion Matrix")
plt.show()

# Tree Visualization
plt.figure(figsize=(12,6))
plot_tree(clf1, feature_names=iris.feature_names,
          class_names=iris.target_names, filled=True)
plt.title("Decision Tree - Iris")
plt.show()



# WINE DATASET

wine = load_wine()
X2, y2 = wine.data, wine.target

X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.3, random_state=42
)

clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X2_train, y2_train)

y2_pred = clf2.predict(X2_test)

print("\nWINE DATASET")
print("Accuracy:", accuracy_score(y2_test, y2_pred))
print("F1 Score (macro):", f1_score(y2_test, y2_pred, average='macro'))

# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y2_test, y2_pred)
plt.title("Wine Confusion Matrix")
plt.show()

# Tree Visualization
plt.figure(figsize=(14,7))
plot_tree(clf2, feature_names=wine.feature_names,
          class_names=wine.target_names, filled=True)
plt.title("Decision Tree - Wine")
plt.show()
```

```python

# FINAL CLEAN DECISION TREE REGRESSION CODE
# Datasets: Diabetes & California Housing


import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes, fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score



#  DIABETES DATASET


diabetes = load_diabetes()
X1, y1 = diabetes.data, diabetes.target

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.3, random_state=42
)

reg1 = DecisionTreeRegressor(max_depth=3, random_state=42)
reg1.fit(X1_train, y1_train)

y1_pred = reg1.predict(X1_test)

print("========== DIABETES DATASET ==========")
print("MAE:", mean_absolute_error(y1_test, y1_pred))
print("RMSE:", np.sqrt(mean_squared_error(y1_test, y1_pred)))
print("R2 Score:", r2_score(y1_test, y1_pred))


# Clean Scatter Plot with Perfect Fit Line
plt.figure(figsize=(6,6))
plt.scatter(y1_test, y1_pred, alpha=0.6)
plt.plot([min(y1_test), max(y1_test)],
         [min(y1_test), max(y1_test)],
         color='red', linewidth=2)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Diabetes: Actual vs Predicted")
plt.grid(True)
plt.show()


# Tree Visualization
plt.figure(figsize=(12,6))
plot_tree(reg1, feature_names=diabetes.feature_names, filled=True)
plt.title("Decision Tree - Diabetes")
plt.show()




#  CALIFORNIA HOUSING DATASET


california = fetch_california_housing()
X2, y2 = california.data, california.target

X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.3, random_state=42
)

reg2 = DecisionTreeRegressor(max_depth=3, random_state=42)
reg2.fit(X2_train, y2_train)

y2_pred = reg2.predict(X2_test)

print("\n========== CALIFORNIA HOUSING DATASET ==========")
print("MAE:", mean_absolute_error(y2_test, y2_pred))
print("RMSE:", np.sqrt(mean_squared_error(y2_test, y2_pred)))
print("R2 Score:", r2_score(y2_test, y2_pred))


# Clean Scatter Plot with Perfect Fit Line
plt.figure(figsize=(6,6))
plt.scatter(y2_test, y2_pred, alpha=0.6)
plt.plot([min(y2_test), max(y2_test)],
         [min(y2_test), max(y2_test)],
         color='red', linewidth=2)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("California Housing: Actual vs Predicted")
plt.grid(True)
plt.show()


# Tree Visualization
plt.figure(figsize=(14,7))
plot_tree(reg2, feature_names=california.feature_names, filled=True)
plt.title("Decision Tree - California Housing")
plt.show()
```
