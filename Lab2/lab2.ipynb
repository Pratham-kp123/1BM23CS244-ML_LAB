**Simple Linear regression**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# np.random.seed(42)
X = 2 * np.random.rand(100, 1)          
y = 4 + 3 * X + np.random.randn(100, 1) 
                                                                                

X
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X_test = np.array([[0], [2]])
y_pred = model.predict(X_test)
y_pred
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```

```python
y_train_pred = model.predict(X)
print("MSE:", mean_squared_error(y, y_train_pred))
print("R² Score:", r2_score(y, y_train_pred))
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, y_train_pred, color='red', label="Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Demo")
plt.show()
```

**Multiple linear regression**

```python

from sklearn.model_selection import train_test_split


```

```python
# 1. Create synthetic dataset
np.random.seed(42)
num_samples = 200

features = np.random.rand(num_samples, 3)
features
```

```python
# True relationship:
# target_value = 5 + 2*f1 + 3*f2 - 1*f3 + noise
true_weights = np.array([2, 3, -1])
target_value = 5 + features @ true_weights + np.random.randn(num_samples) * 0.5
```

```python
# 2. Split dataset
features_train, features_test, target_train, target_test = train_test_split(
    features, target_value, test_size=0.2, random_state=42
)
```

```python

# 3. Train model
regression_model = LinearRegression()
regression_model.fit(features_train, target_train)
```

```python
# 4. Predictions
train_predictions = regression_model.predict(features_train)
test_predictions = regression_model.predict(features_test)
```

```python
# 5. Model parameters
print("Intercept:", regression_model.intercept_)
print("Coefficients:", regression_model.coef_)
```

```python
# 5. Model parameters
print("Intercept:", regression_model.intercept_)
print("Coefficients:", regression_model.coef_)
```

```python
# 6. Evaluation
print("\nTrain R2:", r2_score(target_train, train_predictions))
print("Test R2:", r2_score(target_test, test_predictions))
```

```python
print("\nTrain RMSE:", np.sqrt(mean_squared_error(target_train, train_predictions)))
print("Test RMSE:", np.sqrt(mean_squared_error(target_test, test_predictions)))
```
